```python
from pykrx import stock
from tqdm import tqdm
import pandas as pd
def date_to_str(df) :
    tmp = pd.DataFrame(df.index)
    print(len(tmp))
    for i in range(len(tmp)) :
        tmp.loc[i,'날짜'] = str(tmp.loc[i,'날짜'])[0:4]+str(tmp.loc[i,'날짜'])[5:7]+str(tmp.loc[i,'날짜'])[8:10]
    return tmp
```

```python
# 삼성전자, SK하이닉스로 2001.01.01~2021.10.11까지의 유효거래일 DF 생성
# 5133일로 동일
sk_df = stock.get_market_ohlcv_by_date("20010101", "20211011", "000660") #20211011
sm_df = stock.get_market_ohlcv_by_date("20010101", "20211011", "005930")
sk = date_to_str(sk_df)
sm = date_to_str(sm_df)
date_df = pd.merge(sm, sk, left_on='날짜', right_on='날짜', how='outer')
print(date_df)
```

```python
kd = pd.read_csv ('G:\\공유 드라이브\\WARD\\Stock_Price\\SP_o_kd.csv')
kp = pd.read_csv ('G:\\공유 드라이브\\WARD\\Stock_Price\\SP_o_kp.csv')
```

```python
# KOSDAQ

V_c_sell = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 기관합계
V_c_buy = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 기관합계

V_efn_sell = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 기타금융 (개별)
V_pf_sell = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 연기금 (기관합계 - 나머지?)
V_ec_sell = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 기타법인 (개별)
V_id_sell = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 개인 (개별)
V_fr_sell = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 외국인
V_efr_sell = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 기타외국인

V_efn_buy = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 기타금융
V_pf_buy = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 연기금
V_ec_buy = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 기타법인
V_id_buy = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 개인
V_fr_buy = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 외국인
V_efr_buy = pd.DataFrame(index = kd['티커'], columns = date_df["날짜"]) # 기타외국인

for i in range(len(kd['티커'])):
    for j in tqdm(range(len(date_df["날짜"]))):
        date = date_df["날짜"][j]
        code = kd['티커'][i]
        df2 = stock.get_market_trading_volume_by_date(date, date, str(format(code, '06')), detail=True, on = '매수')
        df3 = stock.get_market_trading_volume_by_date(date, date, str(format(code, '06')), detail=True, on = '매도')
        df4 = stock.get_market_trading_value_by_date(date, date, str(format(code, '06')), on='매수')
        df5 = stock.get_market_trading_value_by_date(date, date, str(format(code, '06')), on='매도')

        V_efn_buy.loc[code][date] = df2.loc[date]['기타금융']
        V_pf_buy.loc[code][date] = df2.loc[date]['연기금']
        V_efr_buy.loc[code][date] = df2.loc[date]['기타외국인']

        V_efn_sell.loc[code][date] = df3.loc[date]['기타금융']
        V_pf_sell.loc[code][date] = df3.loc[date]['연기금']
        V_efr_sell.loc[code][date] = df3.loc[date]['기타외국인']

        V_c_sell.loc[code][date] = df5.loc[date]['기관합계']
        V_ec_sell.loc[code][date] = df5.loc[date]['기타법인']
        V_id_sell.loc[code][date] = df5.loc[date]['개인']
        V_fr_sell.loc[code][date] = df5.loc[date]['외국인합계']
        
        V_c_buy.loc[code][date] = df4.loc[date]['기관합계']
        V_ec_buy.loc[code][date] = df4.loc[date]['기타법인']
        V_id_buy.loc[code][date] = df4.loc[date]['개인']
        V_fr_buy.loc[code][date] = df4.loc[date]['외국인합계']

V_c_sell.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_c_sell.csv', index = True, header = True)
V_c_buy.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_c_buy.csv', index = True, header = True)

V_efn_sell.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_efn_sell_kd.csv', index = True, header = True)
V_pf_sell.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_pf_sell_kd.csv', index = True, header = True)
V_ec_sell.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_ec_sell_kd.csv', index = True, header = True)
V_id_sell.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_id_sell_kd.csv', index = True, header = True)
V_fr_sell.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_fr_sell_kd.csv', index = True, header = True)
V_efr_sell.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_efr_sell_kd.csv', index = True, header = True)

V_efn_buy.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_efn_buy_kd.csv', index = True, header = True)
V_pf_buy.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_pf_buy_kd.csv', index = True, header = True)
V_ec_buy.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_ec_buy_kd.csv', index = True, header = True)
V_id_buy.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_id_buy_kd.csv', index = True, header = True)
V_fr_buy.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_fr_buy_kd.csv', index = True, header = True)
V_efr_sell.to_csv('G:\\공유 드라이브\\WARD\\Stock_Price\\V_efr_sell_kd.csv', index = True, header = True)
```
