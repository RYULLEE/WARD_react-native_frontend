```python
from pykrx import stock
from tqdm import tqdm
import pandas as pd
import FinanceDataReader as fdr

stocks = fdr.StockListing('KRX')['Symbol']

def date_to_str(df) :
    tmp = pd.DataFrame(df.index)
    print(len(tmp))
    for i in range(len(tmp)) :
        tmp.loc[i,'날짜'] = str(tmp.loc[i,'날짜'])[0:4]+str(tmp.loc[i,'날짜'])[5:7]+str(tmp.loc[i,'날짜'])[8:10]
    return tmp

def start_end_date(code) :
    try : 
        print(code)
        df2 = fdr.DataReader(code)
    except : 
        print(code)
        df2 = fdr.DataReader(code, exchange='KRX-DELISTING')
    if len(df2) == 0:
        print('0')
        return 0
    start_date = str(df2.index[0])[0:4]+str(df2.index[0])[5:7]+str(df2.index[0])[8:10]
    end_date = str(df2.index[-1])[0:4]+str(df2.index[-1])[5:7]+str(df2.index[-1])[8:10]
    if int(start_date) < int('20010101') :
        start_date = '20010101'
    return [start_date, end_date]
```

```python
# 삼성전자, SK하이닉스로 2001.01.01~2021.10.11까지의 유효거래일 DF 생성
# 5133일로 동일
sk_df = stock.get_market_ohlcv_by_date("20010101", "20211011", "000660") #20211011
sm_df = stock.get_market_ohlcv_by_date("20010101", "20211011", "005930")
sk = date_to_str(sk_df)
sm = date_to_str(sm_df)
date_df = pd.merge(sm, sk, left_on='날짜', right_on='날짜', how='outer')
date_df_stamp = sk_df.index
print(sk_df.index)
```

```python
kd = pd.read_csv ('G:\\공유 드라이브\\WARD\\Stock_Price\\SP_o_kd.csv', dtype=object)
kp = pd.read_csv ('G:\\공유 드라이브\\WARD\\Stock_Price\\SP_o_kp.csv', dtype=object)

for i in range(len(kd['티커'])) :
    kd.loc[i, '티커'] = str(kd['티커'][i]).zfill(6)
               
for i in range(len(kp['티커'])) :
    kp.loc[i, '티커'] = str(kp['티커'][i]).zfill(6)
    
kp
```

```python
# KOSDAQ - 매수

V_fni_buy =  pd.DataFrame(index = date_df_stamp) # 금융투자
V_ini_buy = pd.DataFrame(index = date_df_stamp)  # 보험
V_it_buy = pd.DataFrame(index = date_df_stamp)   # 투신
V_pe_buy = pd.DataFrame(index = date_df_stamp)   # 사모
V_bk_buy = pd.DataFrame(index = date_df_stamp)   # 은행
V_efn_buy = pd.DataFrame(index = date_df_stamp)  # 기타금융
V_pf_buy = pd.DataFrame(index = date_df_stamp)   # 연기금
V_ec_buy = pd.DataFrame(index = date_df_stamp)   # 기타법인
V_id_buy = pd.DataFrame(index = date_df_stamp)   # 개인
V_fr_buy = pd.DataFrame(index = date_df_stamp)   # 외국인
V_efr_buy = pd.DataFrame(index = date_df_stamp)  # 기타외국인

i = 0
code = kd['티커'][i]
date = start_end_date(code)
if date != 0 :
    try :
        df = stock.get_market_trading_volume_by_date(date[0], date[1], code, detail = True, on = '매수')
        df_fni = df['금융투자'].rename(code)
        df_ini = df['보험'].rename(code)
        df_it = df['투신'].rename(code)
        df_pe = df['사모'].rename(code)
        df_bk = df['은행'].rename(code)
        df_efn = df['기타금융'].rename(code)
        df_pf = df['연기금'].rename(code)
        df_ec = df['기타법인'].rename(code)
        df_id = df['개인'].rename(code)
        df_fr = df['외국인'].rename(code)
        df_efr = df['기타외국인'].rename(code)

        V_fni_buy = pd.merge(V_fni_buy, df_fni , left_on='날짜', right_on='날짜', how='outer')
        V_ini_buy = pd.merge(V_ini_buy, df_ini , left_on='날짜', right_on='날짜', how='outer')
        V_it_buy = pd.merge(V_it_buy, df_it , left_on='날짜', right_on='날짜', how='outer')
        V_pe_buy = pd.merge(V_pe_buy, df_pe , left_on='날짜', right_on='날짜', how='outer')
        V_bk_buy = pd.merge(V_bk_buy, df_bk , left_on='날짜', right_on='날짜', how='outer')
        V_efn_buy = pd.merge(V_efn_buy, df_efn , left_on='날짜', right_on='날짜', how='outer')
        V_pf_buy = pd.merge(V_pf_buy, df_pf , left_on='날짜', right_on='날짜', how='outer')
        V_ec_buy = pd.merge(V_ec_buy, df_ec , left_on='날짜', right_on='날짜', how='outer')
        V_id_buy = pd.merge(V_id_buy, df_id , left_on='날짜', right_on='날짜', how='outer')
        V_fr_buy = pd.merge(V_fr_buy, df_fr , left_on='날짜', right_on='날짜', how='outer')
        V_efr_buy = pd.merge(V_efr_buy, df_efr , left_on='날짜', right_on='날짜', how='outer')
    except :
        print("pass")

for i in tqdm(range(543, len(kd['티커']))):
    code = kd['티커'][i]
    date = start_end_date(code)
    if date == 0 :
        continue
    try :
        df = stock.get_market_trading_volume_by_date(date[0], date[1], code, detail = True, on = '매수')
    except :
        print("pass")

    df_fni = df['금융투자'].rename(code)
    df_ini = df['보험'].rename(code)
    df_it = df['투신'].rename(code)
    df_pe = df['사모'].rename(code)
    df_bk = df['은행'].rename(code)
    df_efn = df['기타금융'].rename(code)
    df_pf = df['연기금'].rename(code)
    df_ec = df['기타법인'].rename(code)
    df_id = df['개인'].rename(code)
    df_fr = df['외국인'].rename(code)
    df_efr = df['기타외국인'].rename(code)

    V_fni_buy = pd.merge(V_fni_buy, df_fni , left_on='날짜', right_on='날짜', how='outer')
    V_ini_buy = pd.merge(V_ini_buy, df_ini , left_on='날짜', right_on='날짜', how='outer')
    V_it_buy = pd.merge(V_it_buy, df_it , left_on='날짜', right_on='날짜', how='outer')
    V_pe_buy = pd.merge(V_pe_buy, df_pe , left_on='날짜', right_on='날짜', how='outer')
    V_bk_buy = pd.merge(V_bk_buy, df_bk , left_on='날짜', right_on='날짜', how='outer')
    V_efn_buy = pd.merge(V_efn_buy, df_efn , left_on='날짜', right_on='날짜', how='outer')
    V_pf_buy = pd.merge(V_pf_buy, df_pf , left_on='날짜', right_on='날짜', how='outer')
    V_ec_buy = pd.merge(V_ec_buy, df_ec , left_on='날짜', right_on='날짜', how='outer')
    V_id_buy = pd.merge(V_id_buy, df_id , left_on='날짜', right_on='날짜', how='outer')
    V_fr_buy = pd.merge(V_fr_buy, df_fr , left_on='날짜', right_on='날짜', how='outer')
    V_efr_buy = pd.merge(V_efr_buy, df_efr , left_on='날짜', right_on='날짜', how='outer')
    
V_fni_buy = V_fni_buy.transpose()
V_ini_buy = V_ini_buy.transpose()
V_it_buy = V_it_buy.transpose()
V_pe_buy = V_pe_buy.transpose()
V_bk_buy = V_bk_buy.transpose()
V_efn_buy = V_efn_buy.transpose()
V_pf_buy = V_pf_buy.transpose()
V_ec_buy = V_ec_buy.transpose()
V_id_buy = V_id_buy.transpose()
V_fr_buy = V_fr_buy.transpose()
V_efr_buy = V_efr_buy.transpose()

V_fni_buy.to_csv('V_fni_buy_kd.csv', index = True, header = True)
V_ini_buy.to_csv('V_ini_buy_kd.csv', index = True, header = True)
V_it_buy.to_csv('V_it_buy_kd.csv', index = True, header = True)
V_pe_buy.to_csv('V_pe_buy_kd.csv', index = True, header = True)
V_bk_buy.to_csv('V_bk_buy_kd.csv', index = True, header = True)
V_efn_buy.to_csv('V_efn_buy_kd.csv', index = True, header = True)
V_pf_buy.to_csv('V_pf_buy_kd.csv', index = True, header = True)
V_ec_buy.to_csv('V_ec_buy_kd.csv', index = True, header = True)
V_id_buy.to_csv('V_id_buy_kd.csv', index = True, header = True)
V_fr_buy.to_csv('V_fr_buy_kd.csv', index = True, header = True)
V_efr_buy.to_csv('V_efr_buy_kd.csv', index = True, header = True)
```
