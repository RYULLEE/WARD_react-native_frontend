```python
from pykrx import stock
from tqdm import tqdm
import pandas as pd
def date_to_str(df) :
    tmp = pd.DataFrame(df.index)
    print(len(tmp))
    for i in range(len(tmp)) :
        tmp.loc[i,'날짜'] = str(tmp.loc[i,'날짜'])[0:4]+str(tmp.loc[i,'날짜'])[5:7]+str(tmp.loc[i,'날짜'])[8:10]
    return tmp
```

```python
# 삼성전자, SK하이닉스로 2001.01.01~2021.10.11까지의 유효거래일 DF 생성
# 5133일로 동일
sk_df = stock.get_market_ohlcv_by_date("20010101", "20211011", "000660") #20211011
sm_df = stock.get_market_ohlcv_by_date("20010101", "20211011", "005930")
sk = date_to_str(sk_df)
sm = date_to_str(sm_df)
date_df = pd.merge(sm, sk, left_on='날짜', right_on='날짜', how='outer')
print(date_df)
```

```python
# KOSPI

market = 'KOSPI'

SP_o = [] # 시가
SP_h = [] # 고가
SP_l = [] # 저가
SP_c = [] # 종가
SP_v = [] # 거래량
SP_cv = [] # 거래대금
SP_dif = [] # 등락률

# 첫 컬럼 설정 및 초기화
date = date_df["날짜"][0]
df = stock.get_market_ohlcv_by_ticker(date, market=market)
df_o = pd.DataFrame(df['시가'])
df_h = pd.DataFrame(df['고가'])
df_l = pd.DataFrame(df['저가'])
df_c = pd.DataFrame(df['종가'])
df_v = pd.DataFrame(df['거래량'])
df_cv = pd.DataFrame(df['거래대금'])
df_dif = pd.DataFrame(df['등락률'])

df_o.columns = [date]
df_h.columns = [date]
df_l.columns = [date]
df_c.columns = [date]
df_v.columns = [date]
df_cv.columns = [date]
df_dif.columns = [date]

SP_o = df_o # 시가
SP_h = df_h # 고가
SP_l = df_l # 저가
SP_c = df_c # 종가
SP_v = df_v # 거래량
SP_cv = df_cv # 거래대금
SP_dif = df_dif # 등락률

# 이후 두번째 부터 자동화

for i in tqdm(range(1, len(date_df))) :
    date = date_df["날짜"][i]
    df = stock.get_market_ohlcv_by_ticker(date, market = market)
    df_o = pd.DataFrame(df['시가'])
    df_h = pd.DataFrame(df['고가'])
    df_l = pd.DataFrame(df['저가'])
    df_c = pd.DataFrame(df['종가'])
    df_v = pd.DataFrame(df['거래량'])
    df_cv = pd.DataFrame(df['거래대금'])
    df_dif = pd.DataFrame(df['등락률'])

    df_o.columns = [date]
    df_h.columns = [date]
    df_l.columns = [date]
    df_c.columns = [date]
    df_v.columns = [date]
    df_cv.columns = [date]
    df_dif.columns = [date]

    SP_o = pd.merge(SP_o, df_o, left_on='티커', right_on='티커', how='outer')
    SP_h = pd.merge(SP_h, df_h, left_on='티커', right_on='티커', how='outer')
    SP_l = pd.merge(SP_l, df_l, left_on='티커', right_on='티커', how='outer')
    SP_c = pd.merge(SP_c, df_c, left_on='티커', right_on='티커', how='outer')
    SP_v = pd.merge(SP_v, df_v, left_on='티커', right_on='티커', how='outer')
    SP_cv = pd.merge(SP_cv, df_cv, left_on='티커', right_on='티커', how='outer')
    SP_dif = pd.merge(SP_dif, df_dif, left_on='티커', right_on='티커', how='outer')
SP_o.to_csv('SP_o_kp.csv', index = True, header = True)
SP_h.to_csv('SP_h_kp.csv', index = True, header = True)
SP_l.to_csv('SP_l_kp.csv', index = True, header = True)
SP_c.to_csv('SP_c_kp.csv', index = True, header = True)
SP_v.to_csv('SP_v_kp.csv', index = True, header = True)
SP_cv.to_csv('SP_cv_kp.csv', index = True, header = True)
SP_dif.to_csv('SP_dif_kp.csv', index = True, header = True)
```

```python
SP_DIV = [] # 배당수익률

# 첫 컬럼 설정 및 초기화
date = date_df["날짜"][0]
df = stock.get_market_fundamental_by_ticker(date, market=market)
df_DIV = pd.DataFrame(df['DIV'])

df_DIV.columns = [date]

SP_DIV = df_DIV # 배당수익률

# 이후 두번째 부터 자동화

for i in tqdm(range(1, len(date_df))) :

    date = date_df["날짜"][i]
    df = stock.get_market_fundamental_by_ticker(date, market=market)
    df_DIV = pd.DataFrame(df['DIV'])

    df_DIV.columns = [date]

SP_DIV.to_csv('SP_DIV_kp.csv', index = True, header = True)
```
